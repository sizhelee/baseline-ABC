```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from tqdm import trange
```

```python
npy_path = "./result/jgl009.npy"
routes = np.load(npy_path)
routes.shape
```

```python
estimator = PCA(n_components=2)
plt.figure(figsize=(10, 10))
colors = np.random.rand(routes.shape[1])
for i in trange(int(routes.shape[0])):
    route = routes[i]
    pca_route = estimator.fit_transform(route)
    
    ax=plt.gca()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.subplot(10, 10, i+1)
    plt.scatter(pca_route.T[0], pca_route.T[1], s=10, c=colors)

ax=plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.suptitle("Trails of Bees")
plt.show()
```
